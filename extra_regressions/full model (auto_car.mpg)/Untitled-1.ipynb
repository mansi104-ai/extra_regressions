```python
#importing the libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

#import linear regression machine learning
from sklearn.linear_model import LinearRegression

#import plotting libraries
import matplotlib.pyplot as plt
import seaborn as sns
import matplotlib.style
plt.style.use('classic')

%matplotlib inline
```

```python
#reading the csv file into pandas dataframe
mpg_df = pd.read_csv('car-mpg.csv')
```

```python
#check the top few records of the data strucutre
mpg_df.head(20)
```

```python
#drop the car name as it is useless for the model
mpg_df = mpg_df.drop('car_name', axis= 1)
```

```python
mpg_df.head()
```

```python
# Replace the numbers in categorical variables with the actual country names in the origin col
mpg_df['origin'] = mpg_df['origin'].replace({1: 'america', 2:'europe', 3:'asia'})
```

```python
mpg_df.head()
```

```python
#convert categorical variable into dummy/indicator variable as many columns will lead to confusio
#this is also known as one hot encoding , the column name will be america , europe and asia with one hot encoding
mpg_df = pd.get_dummies(mpg_df , columns = ["origin"])
```
